##### Creamos una Función que nos Brinda la Cantidad de Votos según el Percentil:

```python
def obtener_valor_por_percentil(ecdf, percentil):
    p = percentil / 100
    return ecdf.inverse(p)
```

```python
print(obtener_valor_por_percentil(ecdf_union, 75))
print(obtener_valor_por_percentil(ecdf_union, 50))
print(obtener_valor_por_percentil(ecdf_union, 25))
```

##### Aplicamos el Código:

```python
number_samples = 1000 #generamos 1000 muestras
sample_size = 35 # tomamos para cada una, 35 pingüinos

samples_df = pd.DataFrame()

np.random.seed(42)
for i in range(1 + number_samples + 1):
    sex_numeric_sample = Processed_coalicion.sample(sample_size, replace=True).to_numpy()
    sample_name = f"sample_{i}"
    samples_df[sample_name] = sex_numeric_sample

coalicion_population_mean = samples_df.mean().mean()
print(f"La población media de candidaturas electas en México por Coalición es del {coalicion_population_mean * 100:.4f}%")
```
